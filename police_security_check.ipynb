```python
create_table_query = """
CREATE TABLE IF NOT EXISTS traffic (
    stop_date DATE,
    stop_time TIME,
    country_name VARCHAR(100),
    driver_gender VARCHAR(50),
    driver_age INT,
    driver_race VARCHAR(50),
    violation_raw VARCHAR(255),
    violation VARCHAR(255),
    search_conducted VARCHAR(50),
    search_type VARCHAR(100),
    stop_outcome VARCHAR(100),
    is_arrested VARCHAR(5),
    stop_duration VARCHAR(50),
    drugs_related_stop VARCHAR(5),
    vehicle_number VARCHAR(50),
    KEY idx_date (stop_date),
    KEY idx_country (country_name),
    KEY idx_violation (violation)
);
"""

mycursor.execute("USE securitycheck;") 
mycursor.execute(create_table_query)
connector.connect()
print("Table 'traffic' created successfully!")
```

```python
mycursor.execute("CREATE DATABASE IF NOT EXISTS traffic;")
print("MySQL database 'traffic' created successfully!")
```

```python
try:
    # Convert DataFrame to list of tuples
    data_list = df.values.tolist()
    
    # Connect to the specific database
    connector = pymysql.connect(
        host="localhost",
        user="root",
        password="1234",
        database="securitycheck"  # Specify database
    )
    
    mycursor = connector.cursor()
    
    query = """
    INSERT INTO traffic (
        stop_date, stop_time, country_name, driver_gender,
        driver_age, driver_race, violation_raw,
        violation, search_conducted, search_type, stop_outcome,
        is_arrested, stop_duration, drugs_related_stop, vehicle_number
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    
    # Insert in batches to prevent memory issues
    batch_size = 1000
    for i in range(0, len(data_list), batch_size):
        batch = data_list[i:i + batch_size]
        mycursor.executemany(query, batch)
        connector.commit()
        print(f"Inserted batch {i//batch_size + 1}, {len(batch)} rows")

    print(f"Total {len(data_list)} rows inserted successfully!")

except pymysql.Error as e:
    print(f"MySQL Error: {e}")
    if connector:
        connector.rollback()
finally:
    if mycursor:
        mycursor.close()
    if connector:
        connector.close()
```
